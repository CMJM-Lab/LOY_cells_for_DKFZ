```r
library(NanoMethViz)
library(bsseq)
library (tibble)
sessionInfo()
```

```r
samples_df <- tibble(
  sample = c("A3","C1","C8","D5","D11","E5B","E6B","F10"),
  group  = c("ROY","ROY","ROY","ROY","LOY","LOY","LOY","LOY")
)
```

```r
bam_dir <- "data/01_bam_files" #data can be provided upon request
```

```r
bam_files <- c(
  A3  = file.path(bam_dir, "A3_sorted.bam"),
  C1  = file.path(bam_dir, "C1_sorted.bam"),
  C8  = file.path(bam_dir, "C8_sorted.bam"),
  D5  = file.path(bam_dir, "D5_sorted.bam"),
  D11 = file.path(bam_dir, "D11_sorted.bam"),
  E5B = file.path(bam_dir, "E5B_sorted.bam"),
  E6B = file.path(bam_dir, "E6B_sorted.bam"),
  F10 = file.path(bam_dir, "F10_sorted.bam")
)
```

```r
setequal(samples_df$sample, names(bam_files))
```

```r
mbf <- ModBamFiles(
  samples = names(bam_files),
  paths   = unname(bam_files)
)
```

```r
exon_tibble <- readRDS(
  "data/hg38_exons_EnsDb_v86.rds" #data can be provided upon request
)
```

```r
head(exon_tibble)
colnames(exon_tibble)
```

```r
exon_tibble$chr <- paste0("chr", exon_tibble$chr)
```

```r
exon_tibble$seqnames <- exon_tibble$chr
```

```r
valid_chr <- c(
  paste0("chr", 1:22),
  "chrX", "chrY", "chrMT"
)
```

```r
exon_tibble <- exon_tibble |>
  dplyr::filter(chr %in% valid_chr)
```

```r
exon_tibble <- exon_tibble |>
  dplyr::filter(chr %in% valid_chr)
```

```r
options("NanoMethViz.site_filter" = 10) #  Removes coverage below 10. 
```

```r
br <- ModBamResult(
  methy   = mbf,          # your ModBamFiles object
  samples = samples_df,   # your sample annotation
  exons   = exon_tibble,
)
```

```r
dmr <- readRDS(
  "data/DMR_LOY_vs_ROY_delta0.2.rds" #data can be provided upon request
)
```

```r
 p <- plot_gene(
  x        = br,
  gene     = "THY1", #change the name of the gene of interest
  anno_regions = dmr,
  heatmap  = TRUE,
  span     = 0.05,
  line_size = 0.5,
  palette  = ggplot2::scale_colour_manual(
    values = c(
      "LOY" = "#0000C0",
      "ROY" = "#ED7D31"
    )
  )
)
p
```
